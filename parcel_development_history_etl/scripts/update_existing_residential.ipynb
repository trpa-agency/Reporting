## 2b. Fix El Dorado County APN Format

El Dorado County changed their APN format by zero-padding the suffix:
- **Old (CSV):**  e.g. 
- **New (FC, later years):**  e.g. 

We detect affected APNs, check which format the FC actually uses per year, and normalize  so the join works correctly.

```python
import re

# Build APN -> COUNTY lookup from the full FC
apn_county = (
    df_fc.dropna(subset=["APN", "COUNTY"])
    .groupby("APN")["COUNTY"]
    .first()
    .to_dict()
)
el_dorado_apns = {apn for apn, c in apn_county.items() if c == "EL"}
print(f"Unique APNs in FC           : {len(apn_county):,}")
print(f"El Dorado (COUNTY=EL) APNs  : {len(el_dorado_apns):,}")

# Show county breakdown
display(
    df_fc.dropna(subset=["APN","COUNTY"])
    .drop_duplicates("APN")[["APN","COUNTY"]]
    .groupby("COUNTY").size()
    .reset_index(name="unique_APNs")
    .sort_values("unique_APNs", ascending=False)
)

# Regex: APN with 2-digit suffix -> candidate for zero-padding
apn_suffix_2digit = re.compile(r"^(\d{3}-\d{2,3})-(\d{2})$")
apn_suffix_3digit = re.compile(r"^(\d{3}-\d{2,3})-0(\d{2})$")

def pad_suffix(apn):
    m = apn_suffix_2digit.match(apn)
    return f"{m.group(1)}-0{m.group(2)}" if m else None

def depad_suffix(apn):
    m = apn_suffix_3digit.match(apn)
    return f"{m.group(1)}-{m.group(2)}" if m else None

EL_DORADO_PAD_YEAR = 2018

# El Dorado APNs in the CSV that have a 2-digit suffix
candidates = {
    apn for apn in df_csv["APN"].unique()
    if apn_suffix_2digit.match(str(apn)) and apn in el_dorado_apns
}
print(f"El Dorado APNs to pad (Year>={EL_DORADO_PAD_YEAR}): {len(candidates):,}")
```

```python
# pad_suffix / depad_suffix / EL_DORADO_PAD_YEAR / candidates defined in c5b

# Case 1: 2-digit suffix -> pad for Year >= EL_DORADO_PAD_YEAR
padded_map = {apn: pad_suffix(apn) for apn in candidates}

# Case 2: CSV APNs already in 3-digit format that need de-padding for Year < EL_DORADO_PAD_YEAR
already_padded = {
    apn for apn in df_csv["APN"].unique()
    if apn_suffix_3digit.match(str(apn)) and apn in el_dorado_apns
}
depadded_map = {apn: depad_suffix(apn) for apn in already_padded}

print(f"APNs to pad   (2-digit->3-digit, Year>={EL_DORADO_PAD_YEAR}): {len(padded_map):,}")
print(f"APNs to depad (3-digit->2-digit, Year<{EL_DORADO_PAD_YEAR}) : {len(depadded_map):,}")

df_csv["APN_original"] = df_csv["APN"].copy()

def fix_apn(row):
    apn, year = row["APN"], row["Year"]
    if apn in padded_map  and year >= EL_DORADO_PAD_YEAR:
        return padded_map[apn]
    if apn in depadded_map and year <  EL_DORADO_PAD_YEAR:
        return depadded_map[apn]
    return apn

df_csv["APN"] = df_csv.apply(fix_apn, axis=1)

changed       = df_csv["APN"] != df_csv["APN_original"]
padded_rows   = changed & (df_csv["Year"] >= EL_DORADO_PAD_YEAR)
depadded_rows = changed & (df_csv["Year"] <  EL_DORADO_PAD_YEAR)

yrs_pad   = sorted(df_csv.loc[padded_rows,   "Year"].unique())
yrs_depad = sorted(df_csv.loc[depadded_rows, "Year"].unique())
print(f"Rows padded    (Year>={EL_DORADO_PAD_YEAR}): {padded_rows.sum():,}  years: {yrs_pad}")
print(f"Rows de-padded (Year<{EL_DORADO_PAD_YEAR}) : {depadded_rows.sum():,}  years: {yrs_depad}")
df_csv[changed][["APN_original","APN","Year"]].sort_values(["APN_original","Year"]).head(20)
```

## 4b. Resolve Missing APN x Year via Centroid Spatial Match

Many "missing" rows are the same physical parcel under a different historical APN.
For each missing CSV APN x Year, find which FC parcel for that year contains
the centroid of the CSV APN geometry -> crosswalk (FC_APN, Year) -> CSV_APN
-> update those FC rows instead of inserting new ones.


```python
# Step 1: get geometry for each missing CSV APN from the FC (latest year it appears)
missing_csv_apns = set(df_missing_apn_year["APN"].unique())
print(f"CSV APNs needing centroid lookup: {len(missing_csv_apns):,}")

csv_apn_geom = {}  # APN -> (SHAPE, source_year)
with arcpy.da.SearchCursor(fc_path, [FC_APN_FIELD, FC_YEAR_FIELD, "SHAPE@"]) as cur:
    for row in cur:
        apn  = str(row[0]).strip() if row[0] else None
        year = int(row[1]) if row[1] is not None else None
        geom = row[2]
        if apn not in missing_csv_apns or geom is None:
            continue
        if apn not in csv_apn_geom or year > csv_apn_geom[apn][1]:
            csv_apn_geom[apn] = (geom, year)

found_geom    = {a for a in missing_csv_apns if a in csv_apn_geom}
still_no_geom = missing_csv_apns - found_geom
print(f"APNs with geometry in FC : {len(found_geom):,}")
print(f"APNs with no FC geometry : {len(still_no_geom):,}  <- MapServer needed")
if still_no_geom:
    print("  Sample:", sorted(still_no_geom)[:10])
```

```python
# Step 2: build in-memory centroid point FC
centroid_fc = "memory/csv_apn_centroids"
if arcpy.Exists(centroid_fc):
    arcpy.management.Delete(centroid_fc)

sr = arcpy.Describe(fc_path).spatialReference
arcpy.management.CreateFeatureclass("memory", "csv_apn_centroids", "POINT", spatial_reference=sr)
arcpy.management.AddField(centroid_fc, "CSV_APN",  "TEXT", field_length=50)
arcpy.management.AddField(centroid_fc, "SRC_YEAR", "LONG")

with arcpy.da.InsertCursor(centroid_fc, ["SHAPE@", "CSV_APN", "SRC_YEAR"]) as cur:
    for apn, (geom, src_year) in csv_apn_geom.items():
        c = geom.centroid
        cur.insertRow([arcpy.PointGeometry(arcpy.Point(c.X, c.Y), sr), apn, src_year])

cnt = int(arcpy.management.GetCount(centroid_fc).getOutput(0))
print(f"Centroid FC: {cnt:,} points")
```

```python
# Step 3: spatial join centroids -> FC parcels per missing year
# Pass 1: WITHIN (centroid inside parcel)
# Pass 2: CLOSEST with distance cap for anything not matched in pass 1
CLOSEST_MAX_METERS = 50  # ignore matches further than this

missing_years  = sorted(df_missing_apn_year["Year"].unique())
crosswalk_rows = []
join_fc = "memory/centroid_join"

for year in missing_years:
    apns_this_year = set(df_missing_apn_year.loc[df_missing_apn_year["Year"] == year, "APN"])
    apns_with_geom = apns_this_year & found_geom
    if not apns_with_geom:
        continue

    apn_list = ",".join([chr(39) + a + chr(39) for a in apns_with_geom])
    fc_lyr   = "fc_yr_lyr"
    pt_lyr   = "pt_yr_lyr"
    arcpy.management.MakeFeatureLayer(fc_path,    fc_lyr, f"{FC_YEAR_FIELD} = {year}")
    arcpy.management.MakeFeatureLayer(centroid_fc, pt_lyr, f"CSV_APN IN ({apn_list})")

    matched_pass1 = set()

    # ── Pass 1: WITHIN ──────────────────────────────────────────────────
    if arcpy.Exists(join_fc): arcpy.management.Delete(join_fc)
    arcpy.analysis.SpatialJoin(
        target_features=pt_lyr, join_features=fc_lyr,
        out_feature_class=join_fc,
        join_operation="JOIN_ONE_TO_ONE", join_type="KEEP_COMMON",
        match_option="WITHIN",
    )
    with arcpy.da.SearchCursor(join_fc, ["CSV_APN", FC_APN_FIELD]) as cur:
        for row in cur:
            if row[0] and row[1]:
                crosswalk_rows.append({"CSV_APN": row[0], "FC_APN": row[1],
                                       "Year": year, "match": "within"})
                matched_pass1.add(row[0])

    # ── Pass 2: CLOSEST for unmatched ───────────────────────────────────
    unmatched = apns_with_geom - matched_pass1
    if unmatched:
        apn_list2 = ",".join([chr(39) + a + chr(39) for a in unmatched])
        pt_lyr2   = "pt_yr_lyr2"
        arcpy.management.MakeFeatureLayer(centroid_fc, pt_lyr2,
                                          f"CSV_APN IN ({apn_list2})")
        if arcpy.Exists(join_fc): arcpy.management.Delete(join_fc)
        arcpy.analysis.SpatialJoin(
            target_features=pt_lyr2, join_features=fc_lyr,
            out_feature_class=join_fc,
            join_operation="JOIN_ONE_TO_ONE", join_type="KEEP_COMMON",
            match_option="CLOSEST",
            search_radius=f"{CLOSEST_MAX_METERS} Meters",
            distance_field_name="DISTANCE",
        )
        p2 = 0
        with arcpy.da.SearchCursor(join_fc, ["CSV_APN", FC_APN_FIELD, "DISTANCE"]) as cur:
            for row in cur:
                if row[0] and row[1] and row[2] <= CLOSEST_MAX_METERS:
                    crosswalk_rows.append({"CSV_APN": row[0], "FC_APN": row[1],
                                           "Year": year, "match": f"closest_{row[2]:.1f}m"})
                    p2 += 1
        if arcpy.Exists(pt_lyr2): arcpy.management.Delete(pt_lyr2)

    for tmp in [fc_lyr, pt_lyr, join_fc]:
        if arcpy.Exists(tmp): arcpy.management.Delete(tmp)

    n_pass1 = len(matched_pass1)
    n_total = len([r for r in crosswalk_rows if r["Year"] == year])
    n_pass2 = n_total - n_pass1
    print(f"  {year}: {len(apns_with_geom):>4} centroids  "
          f"within={n_pass1:>4}  closest={n_pass2:>3}  "
          f"unresolved={len(apns_this_year) - n_total:>3}")

df_crosswalk = pd.DataFrame(crosswalk_rows)
print(f"\nTotal crosswalk rows   : {len(df_crosswalk):,}")
print(f"Unique CSV APNs matched: {df_crosswalk['CSV_APN'].nunique():,}")
print(f"Via WITHIN             : {(df_crosswalk['match']=='within').sum():,}")
print(f"Via CLOSEST            : {(df_crosswalk['match']!='within').sum():,}")
```

```python
# Step 4: review crosswalk and save
df_crosswalk["APN_changed"] = df_crosswalk["CSV_APN"] != df_crosswalk["FC_APN"]

matched_apn_year = set(zip(df_crosswalk["CSV_APN"], df_crosswalk["Year"]))
df_still_missing = df_missing_apn_year[
    ~df_missing_apn_year.apply(lambda r: (r["APN"], r["Year"]) in matched_apn_year, axis=1)
].copy()

print(f"Crosswalk: {len(df_crosswalk):,} matches")
print(f"  Same APN (parcel exists under same id)  : {(~df_crosswalk['APN_changed']).sum():,}")
print(f"  APN renamed (historical match)          : {df_crosswalk['APN_changed'].sum():,}")
print(f"Still truly missing (no spatial match)   : {len(df_still_missing):,}")

# Save crosswalk for review
out_dir = local_path / "data" / "processed_data"
crosswalk_out = out_dir / "ResUnits_APN_crosswalk.csv"
df_crosswalk.sort_values(["Year","CSV_APN"]).to_csv(crosswalk_out, index=False)
print(f"\nSaved crosswalk -> {crosswalk_out}")

print("\nSample renamed parcels (CSV_APN != FC_APN):")
display(df_crosswalk[df_crosswalk["APN_changed"]].head(20))
```

```python
# Step 5: extend csv_lookup with crosswalk so the main update
# assigns the right Residential_Units to historical FC APNs.
#
# csv_lookup keyed by (CSV_APN, Year) -> units
# crosswalk maps (CSV_APN, Year) -> FC_APN
# So we add entries: (FC_APN, Year) -> units for every crosswalk match

csv_lookup = {
    (row["APN"], int(row["Year"])): int(row["Residential_Units_CSV"])
    for _, row in df_csv.iterrows()
}

added = 0
for _, row in df_crosswalk.iterrows():
    fc_apn  = row["FC_APN"]
    csv_apn = row["CSV_APN"]
    year    = row["Year"]
    val     = csv_lookup.get((csv_apn, year))
    if val is not None and (fc_apn, year) not in csv_lookup:
        csv_lookup[(fc_apn, year)] = val
        added += 1

print(f"csv_lookup entries before crosswalk: {len(csv_lookup) - added:,}")
print(f"Entries added via crosswalk        : {added:,}")
print(f"csv_lookup total                   : {len(csv_lookup):,}")

# Update df_missing_apn_year to only truly unresolved rows for MapServer
df_missing_apn_year = df_still_missing.copy()
print(f"\nRemaining for MapServer geometry insert: {len(df_missing_apn_year):,}")
```

## 7. Spatial Attribute Updates

Populate spatial attributes for new/null rows.
Pattern mirrors `CountyParcel_Transform.py` in the ParcelUpdate repo:
1. PARCEL_ACRES, PARCEL_SQFT from geometry
2. WITHIN_TRPA_BNDY, WITHIN_BONUSUNIT_BNDY via SelectByLocation
3. Centroid spatial joins: TOWN_CENTER, TAZ, PLAN, ZONING, REGIONAL_LANDUSE

```python
# ── Spatial join source layers (all confirmed service URLs) ──────────────
SPATIAL_SOURCES = {
    "TRPA_bdy"          : "https://maps.trpa.org/server/rest/services/Boundaries/FeatureServer/4",
    "BonusUnit"         : "https://maps.trpa.org/server/rest/services/Housing/MapServer/8",
    "TownCenter"        : "https://maps.trpa.org/server/rest/services/Boundaries/FeatureServer/1",
    "LocationToTownCtr" : "https://maps.trpa.org/server/rest/services/Planning/FeatureServer/4",
    "TAZ"               : "https://maps.trpa.org/server/rest/services/Transportation_Planning/MapServer/6",
    "LocalPlan"         : "https://maps.trpa.org/server/rest/services/Planning/FeatureServer/2",
    "Zoning"            : "https://maps.trpa.org/server/rest/services/Zoning/MapServer/0",
    "RegionalLandUse"   : "https://maps.trpa.org/server/rest/services/LocalPlan/MapServer/7",
}

# Scratch feature classes (all local)
ParcelPoint        = SCRATCH_GDB + "/PHA_Points"
ParcelPoint_Town   = SCRATCH_GDB + "/PHA_Town"
ParcelPoint_LocTc  = SCRATCH_GDB + "/PHA_LocTC"
ParcelPoint_TAZ    = SCRATCH_GDB + "/PHA_TAZ"
ParcelPoint_Plan   = SCRATCH_GDB + "/PHA_Plan"
ParcelPoint_Zoning = SCRATCH_GDB + "/PHA_Zoning"
ParcelPoint_RLU    = SCRATCH_GDB + "/PHA_RLU"

def field_join_calc(update_fc, key_fields, value_fields,
                    source_fc, source_key_fields, source_value_fields):
    """Transfer values from source_fc to update_fc via APN key dict.
    Adapted from fieldJoinCalc_multikey() in ParcelUpdate/utils.py.
    """
    value_dict = {}
    for row in arcpy.da.SearchCursor(source_fc, source_key_fields + source_value_fields):
        k = tuple(str(v) if v is not None else "" for v in row[:len(source_key_fields)])
        if any(k):
            value_dict[k] = list(row[len(source_key_fields):])
    updated = 0
    with arcpy.da.UpdateCursor(update_fc, key_fields + value_fields) as cur:
        for row in cur:
            k = tuple(str(v) if v is not None else "" for v in row[:len(key_fields)])
            if k in value_dict:
                cur.updateRow(list(row[:len(key_fields)]) + value_dict[k])
                updated += 1
    return updated

print("Spatial join sources (all service URLs):")
for k, v in SPATIAL_SOURCES.items():
    print(f"  {k:<20} {v}")
```

```python
year_list   = ", ".join(str(y) for y in csv_years)
where_years = f"{FC_YEAR_FIELD} IN ({year_list})"
arcpy.management.MakeFeatureLayer(LOCAL_FC, "local_lyr", where_years)
n_scope = int(arcpy.management.GetCount("local_lyr").getOutput(0))
print(f"Rows in scope: {n_scope:,}")

# PARCEL_ACRES and PARCEL_SQFT
print("Calculating PARCEL_ACRES / PARCEL_SQFT...")
n = 0
with arcpy.da.UpdateCursor("local_lyr", ["PARCEL_ACRES", "PARCEL_SQFT", "SHAPE@"]) as cur:
    for row in cur:
        if row[2]:
            row[0] = row[2].getArea("PLANAR", "ACRES")
            row[1] = row[2].getArea("PLANAR", "SquareFeetUS")
            cur.updateRow(row)
            n += 1
print(f"  {n:,} rows")

# Boundary flags
def flag_within(lyr, source_fc, field):
    with arcpy.da.UpdateCursor(lyr, [field]) as cur:
        for row in cur: row[0] = 0; cur.updateRow(row)
    sel = arcpy.management.SelectLayerByLocation(lyr, "INTERSECT", source_fc)
    n = 0
    with arcpy.da.UpdateCursor(sel, [field]) as cur:
        for row in cur: row[0] = 1; cur.updateRow(row); n += 1
    arcpy.management.SelectLayerByAttribute(lyr, "CLEAR_SELECTION")
    return n

n1 = flag_within("local_lyr", SPATIAL_SOURCES["TRPA_bdy"],  "WITHIN_TRPA_BNDY")
n2 = flag_within("local_lyr", SPATIAL_SOURCES["BonusUnit"], "WITHIN_BONUSUNIT_BNDY")
print(f"WITHIN_TRPA_BNDY=1       : {n1:,}")
print(f"WITHIN_BONUSUNIT_BNDY=1  : {n2:,}")
```

```python
# Centroid points (created once, used for all joins)
for fc in [ParcelPoint, ParcelPoint_Town, ParcelPoint_LocTc, ParcelPoint_TAZ,
           ParcelPoint_Plan, ParcelPoint_Zoning, ParcelPoint_RLU]:
    if arcpy.Exists(fc): arcpy.management.Delete(fc)

arcpy.management.FeatureToPoint("local_lyr", ParcelPoint, "INSIDE")
print(f"Centroids: {int(arcpy.management.GetCount(ParcelPoint).getOutput(0)):,}")

def sj_transfer(join_src, out_fc, source_fields, target_fields, label):
    arcpy.analysis.SpatialJoin(
        ParcelPoint, join_src, out_fc,
        "JOIN_ONE_TO_ONE", "KEEP_ALL", match_option="HAVE_THEIR_CENTER_IN")
    n = field_join_calc(
        "local_lyr", ["APN"], target_fields,
        out_fc,      ["APN"], source_fields)
    print(f"  {label:<35} {n:,} rows")

# TOWN_CENTER (name of the town center polygon)
sj_transfer(SPATIAL_SOURCES["TownCenter"],      ParcelPoint_Town,
            source_fields=["NAME"],
            target_fields=["TOWN_CENTER"],
            label="TOWN_CENTER")

# LOCATION_TO_TOWNCENTER (separate Planning service)
sj_transfer(SPATIAL_SOURCES["LocationToTownCtr"], ParcelPoint_LocTc,
            source_fields=["LOCATION_TO_TOWNCENTER"],  # confirm field name in service
            target_fields=["LOCATION_TO_TOWNCENTER"],
            label="LOCATION_TO_TOWNCENTER")

# TAZ
sj_transfer(SPATIAL_SOURCES["TAZ"],             ParcelPoint_TAZ,
            source_fields=["TAZ"],
            target_fields=["TAZ"],
            label="TAZ")

# PLAN_ID + PLAN_NAME
sj_transfer(SPATIAL_SOURCES["LocalPlan"],       ParcelPoint_Plan,
            source_fields=["PLAN_ID", "PLAN_NAME"],
            target_fields=["PLAN_ID", "PLAN_NAME"],
            label="PLAN_ID / PLAN_NAME")

# ZONING_ID + ZONING_DESCRIPTION
sj_transfer(SPATIAL_SOURCES["Zoning"],          ParcelPoint_Zoning,
            source_fields=["ZONING_ID", "ZONING_DESCRIPTION"],
            target_fields=["ZONING_ID", "ZONING_DESCRIPTION"],
            label="ZONING")

# REGIONAL_LANDUSE
sj_transfer(SPATIAL_SOURCES["RegionalLandUse"], ParcelPoint_RLU,
            source_fields=["REGIONAL_LANDUSE"],
            target_fields=["REGIONAL_LANDUSE"],
            label="REGIONAL_LANDUSE")

for fc in [ParcelPoint, ParcelPoint_Town, ParcelPoint_LocTc, ParcelPoint_TAZ,
           ParcelPoint_Plan, ParcelPoint_Zoning, ParcelPoint_RLU]:
    if arcpy.Exists(fc): arcpy.management.Delete(fc)

print("\nAll spatial attribute updates complete on local copy.")
print(f"Local FC ready to review: {LOCAL_FC}")
```

## 8. Fetch Missing Geometry from AllParcels MapServer

Service: `https://maps.trpa.org/server/rest/services/AllParcels/MapServer`

One polygon layer per year (2006-2024). For each missing APN x Year we query
the matching layer. 2025 is not yet in the service - those rows fall back to 2024.

Results are written to `C:/GIS/Scratch.gdb` for review **before** anything touches SDE.

```python
from arcgis.features import FeatureLayer

SERVICE_URL = "https://maps.trpa.org/server/rest/services/AllParcels/MapServer"

# Layer ID per year (from service metadata)
YEAR_LAYER = {
    2024: 32, 2023: 31, 2022: 30, 2021: 29,
    2020: 27, 2019: 22, 2018: 20, 2017: 17, 2016: 18,
    2015:  5, 2014:  6, 2013:  7, 2012:  8,
}
FALLBACK_YEAR = 2024  # use 2024 layer for any year not in dict (e.g. 2025)

def get_layer(year):
    layer_id = YEAR_LAYER.get(year, YEAR_LAYER[FALLBACK_YEAR])
    return FeatureLayer(f"{SERVICE_URL}/{layer_id}")

# Group missing APN x Year by year
missing_by_year = df_missing_apn_year.groupby("Year")["APN"].apply(list).to_dict()

print("Missing APN x Year counts by year:")
for yr, apns in sorted(missing_by_year.items()):
    layer_id = YEAR_LAYER.get(yr, YEAR_LAYER[FALLBACK_YEAR])
    note = "" if yr in YEAR_LAYER else f" (no {yr} layer, using {FALLBACK_YEAR} layer {layer_id})"
    print(f"  {yr}: {len(apns):>5,} APNs  ->  layer {layer_id}{note}")
```

```python
BATCH_SIZE = 50

# All available years sorted — used to build nearest-year search order
AVAILABLE_YEARS = sorted(YEAR_LAYER.keys())

def pad_apn(apn):
    import re
    m = re.match(r"^(\d{3}-\d{2,3})-(\d{2})$", apn)
    return f"{m.group(1)}-0{m.group(2)}" if m else None

def nearest_years(target, available=AVAILABLE_YEARS):
    """Return available years sorted by proximity to target year.
    e.g. target=2013 -> [2013, 2014, 2012, 2015, 2011, ...]
    """
    return sorted(available, key=lambda y: (abs(y - target), y))

def query_layer(layer, apns, skip_pad=None):
    """Query a single layer for a list of APNs.
    Returns {csv_apn: feature}. Tries zero-padded suffix if not found (EL county).
    """
    _skip_pad = skip_pad or set()
    found = {}

    def _fetch(apn_list):
        quoted = ",".join(f"'{a}'" for a in apn_list)
        try:
            return layer.query(
                where=f"APN IN ({quoted})",
                out_fields="APN",
                return_geometry=True
            ).features
        except Exception as e:
            print(f"    Warning: {e}")
            return []

    # Pass 1 — original APN format
    for i in range(0, len(apns), BATCH_SIZE):
        for feat in _fetch(apns[i:i + BATCH_SIZE]):
            fc_apn = feat.attributes.get("APN")
            if fc_apn:
                found[fc_apn] = feat

    # Pass 2 — zero-padded retry for El Dorado APNs not yet found
    padded_to_orig = {pad_apn(a): a for a in apns if pad_apn(a)}
    orig_to_padded = {v: k for k, v in padded_to_orig.items()}
    found_orig = (
        set(found.keys())
        | {padded_to_orig[k] for k in found if k in padded_to_orig}
    )
    need_pad = [
        orig_to_padded[a] for a in apns
        if a not in found_orig and a in orig_to_padded and a not in _skip_pad
    ]
    for i in range(0, len(need_pad), BATCH_SIZE):
        for feat in _fetch(need_pad[i:i + BATCH_SIZE]):
            fc_apn = feat.attributes.get("APN")
            if fc_apn and fc_apn in padded_to_orig:
                orig = padded_to_orig[fc_apn]
                found[orig] = feat
                found[orig]._apn_used = fc_apn

    return found


# ── Main fetch loop with nearest-year fallback ───────────────────────────
geom_rows      = []   # {APN, Year, geom_year, geometry}
apn_format_log = []   # El Dorado padding log
still_missing  = []   # APNs not found in ANY year

# Layer cache so we don't re-instantiate for each fallback
_layer_cache = {}
def get_cached_layer(year):
    if year not in _layer_cache:
        _layer_cache[year] = get_layer(year)
    return _layer_cache[year]

print("Querying AllParcels MapServer (with nearest-year fallback)...")

for target_year, apns in sorted(missing_by_year.items()):
    skip_pad = {a for a in apns if apn_county.get(a, "EL") not in ("EL", None, "")}
    remaining = list(apns)  # APNs still needing geometry
    year_order = nearest_years(target_year)

    for search_year in year_order:
        if not remaining:
            break
        layer  = get_cached_layer(search_year)
        result = query_layer(layer, remaining, skip_pad=skip_pad)

        found_this_round = []
        for csv_apn, feat in result.items():
            fc_apn = getattr(feat, "_apn_used", csv_apn)
            geom_rows.append({
                "APN"           : csv_apn,
                "APN_in_service": fc_apn,
                "Year"          : target_year,
                "geom_year"     : search_year,   # which year's layer the geometry came from
                "geometry"      : feat.geometry,
            })
            if fc_apn != csv_apn:
                apn_format_log.append({"Year": target_year, "geom_year": search_year,
                                       "APN_csv": csv_apn, "APN_service": fc_apn})
            found_this_round.append(csv_apn)

        remaining = [a for a in remaining if a not in found_this_round]

        if result and search_year != target_year:
            print(f"  {target_year}: used {search_year} layer for {len(result):,} APNs")

    if remaining:
        still_missing.extend([(a, target_year) for a in remaining])

df_geom = pd.DataFrame(geom_rows)

print(f"\nGeometry fetched   : {len(df_geom):,} rows")
print(f"Still not found    : {len(still_missing):,} rows")

if len(df_geom):
    fallback_used = df_geom[df_geom["geom_year"] != df_geom["Year"]]
    print(f"Fallback year used : {len(fallback_used):,} rows ({len(fallback_used)/len(df_geom)*100:.1f}%)")

if still_missing:
    df_still_missing = pd.DataFrame(still_missing, columns=["APN", "Year"])
    print("\nAPNs not found in any year:")
    display(df_still_missing)

if apn_format_log:
    df_fmt = pd.DataFrame(apn_format_log)
    print(f"\nEl Dorado padding used for {len(df_fmt):,} rows")

df_geom.head()
```

## 10. QA / QC

Three checks:
1. **Spatial completeness** — every in-scope row has a non-null value (or explicit `N/A`) for each spatial attribute
2. **Residential units totals by year** — FC sum matches CSV sum per year
3. **Duplicate APN x Year** — no (APN, Year) appears more than once in the FC


```python
# ── Read FC back into pandas for QA ─────────────────────────────────────
QA_FIELDS = [
    FC_APN_FIELD, FC_YEAR_FIELD, FC_RES_FIELD,
    "PARCEL_ACRES", "PARCEL_SQFT",
    "WITHIN_TRPA_BNDY", "WITHIN_BONUSUNIT_BNDY",
    "TOWN_CENTER", "LOCATION_TO_TOWNCENTER",
    "TAZ", "PLAN_ID", "PLAN_NAME",
    "ZONING_ID", "ZONING_DESCRIPTION",
    "REGIONAL_LANDUSE",
]

# Only check fields that actually exist in the FC
existing_fields = {f.name for f in arcpy.ListFields(LOCAL_FC)}
qa_read_fields = [f for f in QA_FIELDS if f in existing_fields]
missing_qa_fields = set(QA_FIELDS) - existing_fields
if missing_qa_fields:
    print(f"Warning: QA fields not in FC (skipped): {sorted(missing_qa_fields)}")

year_list_qa = ", ".join(str(y) for y in csv_years)
where_qa = f"{FC_YEAR_FIELD} IN ({year_list_qa})"
rows_qa = []
with arcpy.da.SearchCursor(LOCAL_FC, qa_read_fields, where_qa) as cur:
    for row in cur:
        rows_qa.append(dict(zip(qa_read_fields, row)))

df_qa = pd.DataFrame(rows_qa)
df_qa.rename(columns={FC_APN_FIELD: "APN", FC_YEAR_FIELD: "Year",
                       FC_RES_FIELD: "Residential_Units"}, inplace=True)
print(f"FC rows in scope: {len(df_qa):,}  |  Years: {sorted(df_qa['Year'].unique())}")
```

```python
# ── Check 1: Spatial attribute completeness ──────────────────────────────
SPATIAL_ATTR_FIELDS = [
    "PARCEL_ACRES", "PARCEL_SQFT",
    "WITHIN_TRPA_BNDY", "WITHIN_BONUSUNIT_BNDY",
    "TOWN_CENTER", "LOCATION_TO_TOWNCENTER",
    "TAZ", "PLAN_ID", "PLAN_NAME",
    "ZONING_ID", "ZONING_DESCRIPTION",
    "REGIONAL_LANDUSE",
]

completeness_flags = []
print("Spatial attribute completeness (in-scope rows):\n")
print(f"  {'Field':<30} {'Null/Empty':>10} {'Pct':>7}")
print("  " + "-" * 50)

for field in SPATIAL_ATTR_FIELDS:
    if field not in df_qa.columns:
        continue
    col = df_qa[field]
    # Null OR empty string (but 'N/A' is acceptable)
    if col.dtype == object:
        bad_mask = col.isna() | (col.astype(str).str.strip() == "")
    else:
        bad_mask = col.isna()
    n_bad = bad_mask.sum()
    pct   = 100 * n_bad / len(df_qa) if len(df_qa) else 0
    status = "OK" if n_bad == 0 else "FLAG"
    print(f"  {field:<30} {n_bad:>10,} {pct:>6.1f}%  {status}")
    if n_bad > 0:
        bad_rows = df_qa.loc[bad_mask, ["APN", "Year", field]].copy()
        bad_rows["qa_check"] = f"missing_{field}"
        completeness_flags.append(bad_rows)

df_completeness_flags = pd.concat(completeness_flags, ignore_index=True) if completeness_flags else pd.DataFrame()
print(f"\nTotal completeness flag rows: {len(df_completeness_flags):,}")
```

```python
# ── Check 2: Residential_Units totals by year ────────────────────────────
fc_totals  = df_qa.groupby("Year")["Residential_Units"].sum().rename("FC_Total")
csv_totals = df_csv.groupby("Year")["Residential_Units_CSV"].sum().rename("CSV_Total")

totals_cmp = pd.concat([csv_totals, fc_totals], axis=1).loc[
    sorted(csv_years)].fillna(0)
totals_cmp["Diff"]  = totals_cmp["FC_Total"] - totals_cmp["CSV_Total"]
totals_cmp["Match"] = totals_cmp["Diff"].abs() < 0.01

print("Residential_Units totals by year (CSV vs FC):\n")
print(f"  {'Year':>6}  {'CSV_Total':>12}  {'FC_Total':>12}  {'Diff':>10}  Status")
print("  " + "-" * 58)
for year, row in totals_cmp.iterrows():
    status = "OK" if row["Match"] else "FLAG"
    print(f"  {year:>6}  {row['CSV_Total']:>12,.0f}  {row['FC_Total']:>12,.0f}  "
          f"{row['Diff']:>+10,.0f}  {status}")

df_totals_flags = totals_cmp[~totals_cmp["Match"]].reset_index()
df_totals_flags["qa_check"] = "units_total_mismatch"
print(f"\nYears with unit total mismatch: {len(df_totals_flags)}")
```

```python
# ── Check 3: Duplicate APN x Year ───────────────────────────────────────
dup_mask = df_qa.duplicated(subset=["APN", "Year"], keep=False)
df_dups  = df_qa[dup_mask].sort_values(["APN", "Year"]).copy()
df_dups["qa_check"] = "duplicate_apn_year"

print(f"Duplicate APN x Year rows : {len(df_dups):,}")
print(f"Unique APN x Year combos  : {df_dups[['APN','Year']].drop_duplicates().shape[0]:,}")
if len(df_dups):
    print("\nSample duplicates (up to 20 rows):")
    print(df_dups[["APN", "Year", "Residential_Units"]].head(20).to_string(index=False))
```

```python
# ── Summary + save flag files ────────────────────────────────────────────
qa_out = local_path / "data" / "qa_data"
qa_out.mkdir(exist_ok=True)

all_flags = []

if len(df_completeness_flags):
    path1 = qa_out / "qa_spatial_completeness_flags.csv"
    df_completeness_flags.to_csv(path1, index=False)
    print(f"Spatial completeness flags -> {path1}")
    all_flags.append(df_completeness_flags.assign(qa_section="spatial_completeness"))

if len(df_totals_flags):
    path2 = qa_out / "qa_units_total_flags.csv"
    df_totals_flags.to_csv(path2, index=False)
    print(f"Units total mismatch flags -> {path2}")
    all_flags.append(df_totals_flags.assign(qa_section="units_total"))

if len(df_dups):
    path3 = qa_out / "qa_duplicate_apn_year.csv"
    df_dups.to_csv(path3, index=False)
    print(f"Duplicate APN x Year flags -> {path3}")
    all_flags.append(df_dups.assign(qa_section="duplicate_apn_year"))

if all_flags:
    df_all_flags = pd.concat(all_flags, ignore_index=True)
    path_all = qa_out / "qa_all_flags.csv"
    df_all_flags.to_csv(path_all, index=False)
    print(f"\nCombined flag file         -> {path_all}")
    print(f"Total flag rows            : {len(df_all_flags):,}")
else:
    print("All QA checks passed. No flags.")
```

## 10b. Deduplicate FC — Remove Duplicate APN x Year Rows


```python
# ── Remove duplicate APN x Year rows from the local FC ──────────────────
# Strategy:
#   - Null APN rows: delete outright
#   - Duplicate APN x Year: keep the row with the lowest OID, delete the rest

print('Scanning for duplicates...')

from collections import defaultdict

# Pass 1: collect all OIDs per (APN, Year)
apn_year_oids = defaultdict(list)   # (apn, year) -> [oid, ...]
null_apn_oids = []

with arcpy.da.SearchCursor(LOCAL_FC, ['OID@', FC_APN_FIELD, FC_YEAR_FIELD]) as cur:
    for oid, apn, year in cur:
        if apn is None or str(apn).strip() == '':
            null_apn_oids.append(oid)
        else:
            apn_year_oids[(apn, year)].append(oid)

# OIDs to delete: null APNs + all but the lowest OID for each duplicate group
oids_to_delete = set(null_apn_oids)
dup_groups = {k: v for k, v in apn_year_oids.items() if len(v) > 1}
for (apn, year), oids in dup_groups.items():
    keep = min(oids)
    for oid in oids:
        if oid != keep:
            oids_to_delete.add(oid)

print(f'  Null-APN rows to delete    : {len(null_apn_oids)}')
print(f'  Duplicate APN x Year groups: {len(dup_groups)}')
print(f'  Duplicate rows to delete   : {len(oids_to_delete) - len(null_apn_oids)}')
print(f'  Total rows to delete       : {len(oids_to_delete)}')

if not oids_to_delete:
    print('Nothing to delete.')
else:
    # Pass 2: delete by OID
    deleted = 0
    with arcpy.da.UpdateCursor(LOCAL_FC, ['OID@']) as cur:
        for row in cur:
            if row[0] in oids_to_delete:
                cur.deleteRow()
                deleted += 1
    print(f'Deleted {deleted} rows.')

    # Verify
    remaining = defaultdict(list)
    with arcpy.da.SearchCursor(LOCAL_FC, ['OID@', FC_APN_FIELD, FC_YEAR_FIELD]) as cur:
        for oid, apn, year in cur:
            remaining[(apn, year)].append(oid)
    still_duped = {k: v for k, v in remaining.items() if len(v) > 1}
    null_remaining = [v for (k, v) in remaining.items() if k[0] is None]
    print(f'Verification — duplicate groups remaining : {len(still_duped)}')
    print(f'Verification — null-APN rows remaining    : {len(null_remaining)}')
```

## 10c. Diagnose Unit Deficit — Find Lost CSV APNs


```python
# Read FC back for all csv_years, merge with CSV, find rows where
# CSV has units > 0 but FC has 0 — these are the 'lost' units.

read_fields = [FC_APN_FIELD, FC_YEAR_FIELD, FC_RES_FIELD, FC_COUNTY_FIELD]
existing = {f.name for f in arcpy.ListFields(LOCAL_FC)}
read_fields = [f for f in read_fields if f in existing]

year_list_d = ', '.join(str(y) for y in csv_years)
rows = []
with arcpy.da.SearchCursor(LOCAL_FC, read_fields,
                           f"{FC_YEAR_FIELD} IN ({year_list_d})") as cur:
    for row in cur:
        rows.append(dict(zip(read_fields, row)))
df_fc_qa = pd.DataFrame(rows).rename(columns={
    FC_APN_FIELD: 'APN', FC_YEAR_FIELD: 'Year',
    FC_RES_FIELD: 'FC_Units', FC_COUNTY_FIELD: 'COUNTY'})
df_fc_qa['FC_Units'] = df_fc_qa['FC_Units'].fillna(0)
print(f'FC rows read: {len(df_fc_qa):,}')

# Merge FC onto CSV (left = CSV so we see every CSV row)
df_csv_pos = df_csv[df_csv['Residential_Units_CSV'] > 0].copy()
df_merged = df_csv_pos.merge(
    df_fc_qa[['APN', 'Year', 'FC_Units', 'COUNTY']],
    on=['APN', 'Year'], how='left')

# Lost: CSV > 0, FC = 0 or no match
df_lost = df_merged[
    df_merged['FC_Units'].fillna(0) == 0
].copy()
df_lost['FC_Units'] = df_lost['FC_Units'].fillna(0)

print(f'CSV rows with units > 0    : {len(df_csv_pos):,}')
print(f'Lost rows (FC=0 or no match): {len(df_lost):,}')
print(f'Lost units total            : {df_lost["Residential_Units_CSV"].sum():.0f}')
print()

# Break down by year
print('Lost units by year:')
by_yr = df_lost.groupby('Year').agg(
    rows=('APN','count'),
    units=('Residential_Units_CSV','sum')
)
print(by_yr.to_string())
print()

# Break down by county
if 'COUNTY' in df_lost.columns:
    print('Lost units by county (top 15):')
    by_county = df_lost.groupby('COUNTY').agg(
        rows=('APN','count'), units=('Residential_Units_CSV','sum')
    ).sort_values('units', ascending=False)
    print(by_county.head(15).to_string())
    print()

# Sample the lost rows — look at APN patterns
print('Sample lost rows (up to 40, sorted by units desc):')
cols = ['APN', 'Year', 'Residential_Units_CSV', 'FC_Units']
if 'COUNTY' in df_lost.columns:
    cols.insert(2, 'COUNTY')
print(df_lost.nlargest(40, 'Residential_Units_CSV')[cols].to_string(index=False))
print()

# Save for review
out_path = local_path / 'data' / 'qa_data' / 'qa_lost_units.csv'
df_lost.to_csv(out_path, index=False)
print(f'Saved to {out_path}')
```
